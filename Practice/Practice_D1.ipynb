```python
import pandas as pd 
import seaborn as sns
```

```python
data = {
    'Area': [1500, 1800, 2000, 1200, 2200, 1700, 2500, 1300, 2100, 1600, 
             2300, 1900, 1400, 1800, 2400, 2700, 3000, 1600, 2000, 2200],
    'Bedrooms': [3, 4, 3, 2, 5, 3, 4, 2, 4, 3, 
                 4, 3, 2, 3, 4, 5, 6, 3, 3, 4],
    'Location': ['Urban', 'Suburban', 'Urban', 'Rural', 'Suburban', 'Urban', 'Urban', 'Rural', 'Suburban', 'Urban', 
                 'Suburban', 'Rural', 'Rural', 'Urban', 'Suburban', 'Urban', 'Suburban', 'Urban', 'Rural', 'Urban'],
    'Price': [500000, 550000, 600000, 300000, 650000, 530000, 750000, 320000, 700000, 540000, 
              680000, 390000, 310000, 560000, 720000, 800000, 850000, 500000, 600000, 750000],
    'Garage': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes',
               'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes'],
    'Garden': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No',
               'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Year_Built': [2005, 2010, 2000, 1995, 2015, 2008, 2020, 1990, 2018, 2003,
                   2012, 1998, 1992, 2006, 2019, 2021, 2015, 2009, 2013, 2007],
    'Pool': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 
             'Yes', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df['Location'].unique()
```

for LOCATION - OrdinalEncoder
For others - LabelEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
# Initialize OrdinalEncoder with categories in the specified order
oe = OrdinalEncoder(categories=[['Rural', 'Suburban', 'Urban']])  # Rural < Suburban < Urban

# Fit the encoder on the Location column and transform the data
df['Location'] = oe.fit_transform(df[['Location']])

# Display the transformed dataframe
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Garage'] = le.fit_transform(df[['Garage']])
df['Garden'] = le.fit_transform(df[['Garden']])
df['Pool'] = le.fit_transform(df[['Pool']])

df.head()
```

```python
numerical_feature = ['Area','Bedrooms','Price']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[numerical_feature] = scaler.fit_transform(df[numerical_feature])

df.head()
```

```python
sns.pairplot(df)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Price', axis=1), df['Price'], test_size=0.3, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

lr =LinearRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
