```python
import pandas as pd 
```

```python
df = pd.read_csv('Collated_Construction_Information_2.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['S/N','Project Description','Project','Preliminary Cost/m2','Contractor Grade',
                            'Ground Floor Area(m2)','Preliminary Cost Estimate(=N=)','Site Area(m2)',
                            'Planned Project Duration (weeks)','Initial Contract Sum','Unnamed: 20'], axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Project End Date']  =pd.to_datetime(df['Project End Date'], dayfirst=True)
df['Project Start Date']  =pd.to_datetime(df['Project Start Date'], dayfirst=True)
```

```python
df
```

```python
df = df.drop(['Project Start Date', 'Project End Date'], axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Actual Project Cost at Completion']), df['Actual Project Cost at Completion'], test_size=0.2, random_state=42)

```

```python
X_train
```

```python
X_train.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

```

```python
numerical_features = X_train.select_dtypes(include=['number']).columns.tolist()
```

```python
preprocessor = ColumnTransformer([
    ('fill_values', SimpleImputer(), numerical_features),
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'),['Project Type']),
], remainder='passthrough')
```

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Transform X_train using only the preprocessor
X_train_transformed = pipeline.named_steps['preprocessor'].transform(X_train)
categorical_features = ['Project Type']
# Get One-Hot Encoded feature names
ohe_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['ohe'].get_feature_names_out(categorical_features)

# Get all column names (numerical + encoded categorical + passthrough)
feature_names = numerical_features + list(ohe_feature_names)  

# Convert transformed data to DataFrame
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=feature_names, index=X_train.index)
```
