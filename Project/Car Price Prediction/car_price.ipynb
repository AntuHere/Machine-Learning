```python
from sklearn.compose import  ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
preprocessor = ColumnTransformer([
    ('fueltype', OrdinalEncoder(), ['fueltype']),
    ('aspiration', OrdinalEncoder(), ['aspiration']),
    ('doornumber', OrdinalEncoder(), ['doornumber']),
    ('carbody', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['carbody']),
    ('drivewheel', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['drivewheel']),
    ('enginelocation', OrdinalEncoder(), ['enginelocation']),
    ('enginetype', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['enginetype']),
    ('cylindernumber', OrdinalEncoder(categories=[['two', 'three', 'four', 'five', 'six', 'eight', 'twelve']]), ['cylindernumber']),
    ('fuelsystem', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['fuelsystem'])
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['price']), df['price'], test_size=0.2)
```

```python
X_train.head()
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
y_pred
```

```python
y_test.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
MAE = mean_absolute_error(y_test, y_pred,multioutput='uniform_average')
MAE
```

```python
MSE = mean_squared_error(y_test, y_pred,multioutput='uniform_average')
MSE
```

```python
r2_score = r2_score(y_test, y_pred,multioutput='uniform_average')
r2_score
```

```python
print(f"RÂ² Score: {r2_score:.4f}")  # Higher is better (max 1)
print(f"MAE: {MAE:.4f}")       # Lower is better
print(f"MSE: {MSE:.4f}")       # Lower is better
  
```
