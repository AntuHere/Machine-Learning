```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.scatter(df['area'], df['price'], marker='+', color="red")
plt.xlabel('Area')
plt.ylabel('Prices')
plt.show()
```

```python
X = df.iloc[:, 0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict(X.iloc[0].values.reshape(1,1))
```

```python
lr.predict([[3300]])
```

```python
plt.scatter(df['area'], df['price'], marker='+', color="red")
plt.plot(df['area'],lr.predict(X))
plt.xlabel('Area')
plt.ylabel('Prices')
plt.show()
```

```python
predicted_area = pd.read_csv('areas.csv')
predicted_area
```

```python
prices = lr.predict(predicted_area)
```

```python
predicted_area['prices'] = prices
```
