```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data Mining and Data Analyze

Data Visualize

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.title('Per Capital income in Canada')
plt.xlabel('Year')
plt.ylabel('canada_per_capita_income')

plt.show()
```

Data extract for training and testing

```python
X = df.iloc[:, 0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train) #training

lr.predict(X_test) #predicted data
```

```python
y_test
```

```python
lr.predict([[2020]])
```

```python
m = lr.coef_
b = lr.intercept_

p = m * 2020 + b
p
```

```python
year_2020 = [[2020]]  # 2D input for prediction
predicted_income_2020 = lr.predict(year_2020)[0]
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.scatter(2020, predicted_income_2020, color='green', marker='o', s=100, label="Prediction (2020)")
plt.annotate(f"2020: ${predicted_income_2020:,.2f}",  
             (2020, predicted_income_2020),  
             textcoords="offset points",  
             xytext=(20,10),  
             ha='center', fontsize=12, color='green')
plt.title('Per Capital income in Canada')
plt.xlabel('Year')
plt.ylabel('canada_per_capita_income')

plt.show()
```
